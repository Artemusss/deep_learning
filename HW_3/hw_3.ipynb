```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoding_o_h = OneHotEncoder(sparse_output=False, drop='if_binary')

cols_o_h= encoding_o_h.fit_transform(data[cat_cols])

one_hot_df = pd.DataFrame(cols_o_h, columns=encoding_o_h.get_feature_names_out(cat_cols), index=data.index
                          )

data_encoded = pd.concat([data, one_hot_df], axis=1)

# Drop the original categorical columns
data_encoded = data_encoded.drop(cat_cols, axis=1)

data_encoded.info()
```

```python

cat_bin_cols = ['Sex_Male', 'IsSeniorCitizen_1', 'HasPartner_Yes','HasChild_Yes','HasPhoneService_Yes', 'IsBillingPaperless_Yes']
cat_not_bin_cols = ['HasMultiplePhoneNumbers_No', 'HasMultiplePhoneNumbers_No phone service', 'HasMultiplePhoneNumbers_Yes',\
                     'HasInternetService_DSL', 'HasInternetService_Fiber optic', 'HasInternetService_No',\
                          'HasOnlineSecurityService_No', 'HasOnlineSecurityService_No internet service',\
                            'HasOnlineSecurityService_Yes', 'HasOnlineBackup_No', 'HasOnlineBackup_No internet service',\
                                'HasOnlineBackup_Yes', 'HasDeviceProtection_No', 'HasDeviceProtection_No internet service',\
                                    'HasDeviceProtection_Yes', 'HasTechSupportAccess_No', 'HasTechSupportAccess_No internet service',\
                                        'HasTechSupportAccess_Yes', 'HasOnlineTV_No', 'HasOnlineTV_No internet service',\
                                            'HasOnlineTV_Yes', 'HasMovieSubscription_No', 'HasMovieSubscription_No internet service',\
                                                'HasMovieSubscription_Yes', 'HasContractPhone_Month-to-month', 'HasContractPhone_One year',\
                                                    'HasContractPhone_Two year', 'PaymentMethod_Bank transfer (automatic)',\
                                                        'PaymentMethod_Credit card (automatic)', 'PaymentMethod_Electronic check',\
                                                            'PaymentMethod_Mailed check']

fig1, axs1 = plt.subplots(len(cat_bin_cols), 2, figsize=(15, 30))
for i in range(len(cat_bin_cols)):
    col_0 = data_encoded['Churn'][data_encoded[cat_bin_cols[i]] == 0].value_counts()
    col_1 = data_encoded['Churn'][data_encoded[cat_bin_cols[i]] == 1].value_counts()

    axs1[i, 0].pie(col_0, labels=col_0.index, autopct='%1.1f%%', radius=1)
    str1 = cat_bin_cols[i] + ' = 0'
    axs1[i, 0].set_title(str1)

    axs1[i, 1].pie(col_1, labels=col_1.index, autopct='%1.1f%%', radius=1)
    str2 = cat_bin_cols[i] + ' = 1'
    axs1[i, 1].set_title(str2)

fig1.suptitle('Analysis of binary features', fontsize=30)
```

```python
fig2, axs2 = plt.subplots(len(cat_not_bin_cols) // 3 - 1, 3, figsize=(20, 60))

for j in range(len(cat_not_bin_cols) // 3 - 1):
    col_0 = data_encoded['Churn'][data_encoded[cat_not_bin_cols[j]] == 1].value_counts()
    col_1 = data_encoded['Churn'][data_encoded[cat_not_bin_cols[j + 1]] == 1].value_counts()
    col_2 = data_encoded['Churn'][data_encoded[cat_not_bin_cols[j + 2]] == 1].value_counts()

    axs2[j, 0].pie(col_0, labels=col_0.index, autopct='%1.1f%%', radius=1)
    str1 = cat_not_bin_cols[j] + ' = 1'
    axs2[j, 0].set_title(str1)

    axs2[j, 1].pie(col_1, labels=col_1.index, autopct='%1.1f%%', radius=1)
    str2 = cat_not_bin_cols[j + 1] + ' = 1'
    axs2[j, 1].set_title(str2)

    axs2[j, 2].pie(col_2, labels=col_2.index, autopct='%1.1f%%', radius=1)
    str3 = cat_not_bin_cols[j + 2] + ' = 1'
    axs2[j, 2].set_title(str3)

fig2.suptitle('Analysis of binary features', fontsize=30)

fig3, axs3 = plt.subplots(1, 4, figsize=(20, 10))

for k in range(4):
    col = data_encoded['Churn'][data_encoded[cat_not_bin_cols[-k - 1]] == 1].value_counts()
    axs3[k].pie(col, labels=col.index, autopct='%1.1f%%', radius=1)
    str = cat_not_bin_cols[-k - 1] + ' = 1'
    axs3[k].set_title(str)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score
```

   train  test.

```python
X = data_encoded.drop(columns=['Churn'])
y = data_encoded['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

    standart scaler, robust scaler  min max scaler,   . 

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
scaler = RobustScaler()
scaler.fit(X_train)

X_train_rob = scaler.transform(X_train)
X_test_rob = scaler.transform(X_test)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_min_max = scaler.transform(X_train)
X_test_min_max = scaler.transform(X_test)
```

 logistic regression CV.

```python
log_reg = LogisticRegressionCV(scoring='roc_auc', refit=True)

log_reg.fit(X_train_std, y_train)

y_pred_std = log_reg.predict_proba(X_test_std)[:, 1]
print('Result(Standart scaler):', roc_auc_score(y_test, y_pred_std), '\n')

log_reg.fit(X_train_rob, y_train)

y_pred_rob = log_reg.predict_proba(X_test_rob)[:, 1]
print('Result(Robust scaler):', roc_auc_score(y_test, y_pred_rob), '\n')

log_reg.fit(X_train_min_max, y_train)

y_pred_min_max = log_reg.predict_proba(X_test_min_max)[:, 1]
print('Result(MinMax scaler):', roc_auc_score(y_test, y_pred_min_max), '\n')

print('Best value C is', log_reg.C_)
```

```python
from catboost import CatBoostClassifier

X = data.drop(columns=['Churn'])
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

cat_cls = CatBoostClassifier(verbose=False) 

cat_cls.fit(X_train, y_train, cat_features=cat_cols)

y_pred = cat_cls.predict_proba(X_test)[:, 1]

print('Result of cat Boost with default param:', roc_auc_score(y_test, y_pred), '\n')
```

 CatboostClassifier   .

```python
num_tr = [1, 10, 100, 500, 1000, 5000]
l_r = [0.1, 0.01, 0.001, 0.0001, 0.00001]

for i in range(len(num_tr)):
    for j in range(len(l_r)):
        cat_cls = CatBoostClassifier(verbose=False, learning_rate=l_r[j], num_trees=num_tr[i]) 

        cat_cls.fit(X_train, y_train, cat_features=cat_cols)

        y_pred = cat_cls.predict_proba(X_test)[:, 1]

        print('Result of cat Boost with param', 'learning_rate =', l_r[j], 'num_trees =', num_tr[i],\
              ':', roc_auc_score(y_test, y_pred), '\n')
        
```
