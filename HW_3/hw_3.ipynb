```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score
```

Разделим данные на train и test.

```python
X = data_encoded.drop(columns=['Churn'])
y = data_encoded['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

Отнормируем числовые признаки используя standart scaler, robust scaler и min max scaler, сравним в конце. 

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
scaler = RobustScaler()
scaler.fit(X_train)

X_train_rob = scaler.transform(X_train)
X_test_rob = scaler.transform(X_test)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_min_max = scaler.transform(X_train)
X_test_min_max = scaler.transform(X_test)
```

Обучим logistic regression CV.

```python
log_reg = LogisticRegressionCV(scoring='roc_auc', refit=True)

log_reg.fit(X_train_std, y_train)

y_pred_std = log_reg.predict_proba(X_test_std)[:, 1]
print('Result(Standart scaler):', roc_auc_score(y_test, y_pred_std), '\n')

log_reg.fit(X_train_rob, y_train)

y_pred_rob = log_reg.predict_proba(X_test_rob)[:, 1]
print('Result(Robust scaler):', roc_auc_score(y_test, y_pred_rob), '\n')

log_reg.fit(X_train_min_max, y_train)

y_pred_min_max = log_reg.predict_proba(X_test_min_max)[:, 1]
print('Result(MinMax scaler):', roc_auc_score(y_test, y_pred_min_max), '\n')

print('Best value C is', log_reg.C_)
```

Как мы видим результаты особо не различаются, но наибольший при использовании Standart scaler.
